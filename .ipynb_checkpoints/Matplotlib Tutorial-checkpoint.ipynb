# Matplotlib Tutorial


### Anatomy

* Figure :  Whole area chosen for plotting
* Axes   :  Area where data is plotted
* Axis   :  X and Y axis and its limits
* Artist :  Any element on the figure

```python
import math
import matplotlib.pyplot as plt
```

### Creating a figure object

* Figure alone is not going to show anything, we need to add axis atlease
* Default size of figure is 6 inch width and 4 inch height

```python
fig = plt.figure()
```

### Axes

* An axes is the region of the figure, available for plotting data
* An axes object associated with only one figure, but fight can have many axes object
* The axes object can be added to figure using <i>fig.add_subplot()</i> method
* ax = fig.add_subplot(111) or fig.add_subplot(1,1,1) unless nrows and ncols are less than 10

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

### Adding extra parameters to axes object

```python
size = (8,4)
fig = plt.figure(figsize=size)
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.set(
    title='Figure 1',
    xlabel = 'x',
    ylabel ='y',
    xlim = (0,1),
    ylim = (-1,1)
)
ax2.set(
    title='Figure 2',
    xlabel = 'x',
    ylabel ='y',
    xlim = (0,1),
    ylim = (1,10)
)
plt.show()
```

### Setting the additional attributes can alos be done using set_ paramter method

```python
size = (8,4)
fig = plt.figure(figsize = size)
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('Figure 1')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim((0, 1))
ax1.set_ylim((-1, 1))
plt.show()
```

### Introducing plot functions to axes

```python
size = (8, 4)
x = list(range(1, 10))
y1 = [math.sin(i) for i in x]
y2 = [math.log(i) for i in x]

fig = plt.figure(figsize = size)
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.set_title('Figure 1 - sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.plot(x, y1)

ax2.set_title('Figure 2 - log(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('log(x)')
ax2.plot(x, y2)

plt.show()
```

### Introducing legends to the chart

* Legends can be added by settting label parameter and by calling <i>plt.legend()</i>
* Legends position can be controled by passing loc parameter to legend(), eg: upper left, left, right, center right etc,, default: best (finds the best position with minimal overlap)
* Legends can also be moved outside of the figure using bbox_to_anchor paramter

```python
size = (6, 4)
x = list(range(1, 10))
y1 = [math.sin(i) for i in x]
y2 = [math.log(i) for i in x]

fig = plt.figure(figsize = size)
ax1 = fig.add_subplot(1, 1, 1)

ax1.set_title('x vs log(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('log(x)')
ax1.plot(x, y2, label='log(x)')

plt.legend()
plt.show()
```

```python
size = (6, 4)
x = list(range(1, 10))
y1 = [math.sin(i) for i in x]
y2 = [math.log(i) for i in x]

fig = plt.figure(figsize = size)
ax1 = fig.add_subplot(1, 1, 1)

ax1.set_title('x vs log(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('log(x)')
ax1.plot(x, y2, label='log(x)')

plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
plt.show()
```

### Additional parameters to plot function

* we can pass color, linestyle, linewidth, marker, markersize, label to <i>plot()</i>
* linestyle - solid, dashed, none
* color - red, blue, green
* marker - o, >

```python
size = (6, 4)
x = list(range(1, 10))
y = [math.log(i) for i in x]

fig = plt.figure(figsize = size)
ax1 = fig.add_subplot(1, 1, 1)

ax1.set_title('x vs log(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('log(x)')
ax1.plot(x, y, label='log(x)', color='green', marker='>', linestyle='none')

plt.legend()
plt.show()
```

### Apply styles in matplotlib

* we can list all the syles using plt.style.available attribute
* we can apply it using plt.style.use() or plt.style.context() using with statement

```python
plt.style.available
```

```python
plt.style.use('ggplot')

size = (6, 4)
x = list(range(1, 10))
y = [math.log(i) for i in x]

fig = plt.figure(figsize = size)
ax1 = fig.add_subplot(1, 1, 1)

ax1.set_title('x vs log(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('log(x)')
ax1.plot(x, y, label='log(x)', color='green', linestyle='dashed')

plt.legend()
plt.show()
```

```python
size = (6, 4)
x = list(range(1, 10))
y = [math.log(i) for i in x]

with plt.style.context('ggplot'):
    fig = plt.figure(figsize = size)
    ax1 = fig.add_subplot(1, 1, 1)

    ax1.set_title('x vs log(x)')
    ax1.set_xlabel('x')
    ax1.set_ylabel('log(x)')
    ax1.plot(x, y, label='log(x)', color='green', linestyle='dashed')

    plt.legend()
    plt.show()
```

### Multiple axes to a figure using subplot, add_subplot and subplots

* add_subplot is used with fig object
* subplot, subplots are used with plt object

```python
fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(2, 2, 1, title='Figure1')
ax2 = fig.add_subplot(2, 2, 2, title='Figure2')
ax3 = fig.add_subplot(2, 2, 3, title='Figure3')
ax4 = fig.add_subplot(2, 2, 4, title='Figure4')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax1 = plt.subplot(2, 2, 1, title='Figure1')
ax2 = plt.subplot(2, 2, 2, title='Figure2')
ax3 = plt.subplot(2, 2, 3, title='Figure3')
ax4 = plt.subplot(2, 2, 4, title='Figure4')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax1 = plt.subplot(2, 2, (1,2), title='Figure1')
ax3 = plt.subplot(2, 2, 3, title='Figure3')
ax4 = plt.subplot(2, 2, 4, title='Figure4')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax1 = plt.subplot(2, 2, (1,3), title='Figure1')
ax3 = plt.subplot(2, 2, 2, title='Figure3')
ax4 = plt.subplot(2, 2, 4, title='Figure4')
plt.show()
```
